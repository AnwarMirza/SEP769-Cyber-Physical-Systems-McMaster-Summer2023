# $\color{green}{\text{1. Linear Activation Function / Linear Neuron}}$
If $x$ is the input to a linear neuron, the output is given by \\
$$f(x) = x$$

```python
def linearNeuron(x):
  return (x)

x = np.linspace(-10, 10, 200)
yLinear = linearNeuron(x)

fig = plt.figure(figsize=(6,6))
plt.plot(x, yLinear, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.title('Linear Activation Function $f(x)=x$')
plt.grid()
plt.show()
```

# $\color{green}{\text{2. Binary (Heaviside) Step Function}}$
If $x$ is the input to a binary step function neuron, the output is given by

$$ f(x) =  \begin{array}{|cc}
1 & \quad x \ge 0 \\\ 0 & \quad x \lt 0
\end{array}
$$

```python
def stepFunc(x):
  return (0.0 if x < 0.0 else 1.0)

x = np.linspace(-10, 10, 100)
yBinaryStep = [stepFunc(a) for a in x]

fig = plt.figure(figsize=(6,6))
plt.plot(x, yBinaryStep, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.grid()
plt.show()
```

# $\color{green}{\text{2. Bipolar Step Function}}$
If $x$ is the input to a bipolar step function neuron, the output is given by

$$ f(x) =  \begin{array}{|rr}
1 & \quad x \ge 0 \\\ -1 & \quad x \lt 0
\end{array}
$$

```python
def bipolarStepFunc(x):
  return (-1.0 if x < 0.0 else 1.0)

x = np.linspace(-10, 10, 100)
yBipolarStep = [bipolarStepFunc(a) for a in x]

fig = plt.figure(figsize=(6,6))
plt.plot(x, yBipolarStep, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.grid()
plt.show()
```

# $\color{green}{\text{Sigmoid Function}}$
If $x$ is the input to a sigmoid neuron, the output is given by \\
$$f(x) = \frac{1}{1+e^{-x}}$$
This is also known as a $\color{blue}{\text{logistic function}}$ or $\color{blue}{\text{binary sigmoid function}}$ .

```python
def sigmoid(x):
  sigma = 1.0 # steepness paramter
  return (1/(1+np.exp(-sigma*x)))

def stepFunc(x):
  return (0.0 if x < 0.0 else 1.0)

x = np.linspace(-10, 10, 200)
yBinaryStep = [stepFunc(a) for a in x]
ySigmoid = sigmoid(x)

fig = plt.figure(figsize=(6,6))
plt.plot( x, yBinaryStep, '-', linewidth=3, label='Binary Step')
plt.plot( x, ySigmoid, '-', linewidth=3, label='Binary Sigmoid')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# $\color{green}{\text{Bipolar Sigmoid Function}}$
If $x$ is the input to a bipolar sigmoid neuron, the output is given by \\
$$f(x) = \frac{1 - e^{-x}}{1+e^{-x}}$$


```python
def bipolarSigmoid(x):
  k = 1.0 # steepness parameter
  return ((1 -  np.exp(-k*x))/(1+np.exp(-k*x)))

x = np.linspace(-10, 10, 200)
yBSigmoid = bipolarSigmoid(x)

fig = plt.figure(figsize=(6,6))
plt.plot(x, yBSigmoid, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.grid()
plt.show()
```

# $\color{green}{\text{Bipolar Sigmoid Function With Different Steepness}}$
If $x$ is the input to a bipolar sigmoid neuron, the output is given by \\
$$f(x) = \frac{1 - e^{-\sigma x}}{1+e^{-\sigma x}}$$


```python
def bipolarSigmaSigmoid(x, sigma):
  return ((1 -  np.exp(-sigma*x))/(1+np.exp(-sigma*x)))

x = np.linspace(-10, 10, 200)
yBSigmoid1 = bipolarSigmaSigmoid(x, 1.0)
yBSigmoid2 = bipolarSigmaSigmoid(x, 2.0)
yBSigmoid10 = bipolarSigmaSigmoid(x, 10.0)

fig = plt.figure(figsize=(6,6))
plt.plot( x, yBSigmoid1, '-', linewidth=3, label='$\sigma = 1$')
plt.plot( x, yBSigmoid2, '-', linewidth=3, label='$\sigma = 2$')
plt.plot( x, yBSigmoid10, '-', linewidth=3, label='$\sigma = 10$')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

# $\color{green}{\text{Hyperbolic Tangent Function}}$
If $x$ is the input to a hyperbolic tangent neuron, the output is given by \\
$$f(x) = \text{tanh}(x) = \frac{1 - e^{-2x}}{1+e^{-2x}}$$

```python
def tangentHyperbolic(x):
  return (np.tanh(x))

x = np.linspace(-10, 10, 200)
yTanHyp = tangentHyperbolic(x)

fig = plt.figure(figsize=(6,6))
plt.plot(x, yTanHyp, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.grid()
plt.show()
```

# $\color{green}{\text{Rectified Linear Unit (ReLU)}}$
If $x$ is the input to a ReLU neuron, the output is given by
$$f(x) = x^+ = max(0, x)$$
This is also known as a $\color{blue}{\text{ramp function}}$ and known as $\color{blue}{\text{rectifier}}$ in electrical / electronics engineering. \\
Compared to most commonly used logistic sigmoid function, it was found in 2011 that ReLU enables better training of deep neural networks. It has found applications in computer vision, speech recognition and pattern recognition for deep neural networks. It is one of the most commonly used activation function for deep neural networks.\\
The derivative of the rectifier is the binary step (Heaviside) function (assuming the derivative to be 1 at $x=0$):
$$f^{\prime}(x) = \begin{array}{|cc} \quad 1 & x \ge 0 \\ \quad 0 & x < 0 \end{array}$$

```python
def relu(x):
  return np.maximum(0.0, x)

xVal = np.linspace(-3, 3, 200)
yRelu = [relu(x) for x in xVal]

fig = plt.figure(figsize=(6,6))
plt.plot(xVal, yRelu, '-', linewidth=3)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x) = ReLU(x)$', fontsize=18)
plt.grid()
plt.show()
```

# $\color{green}{\text{Leaky ReLU}}$
If $x$ is the input to a leaky ReLU neuron, the output is given by \\
$$f(x) = \begin{array}{|ll}
\ x & \quad x \gt 0 \\ \ 0.01x & \quad \text{otherwise} \end{array}$$
One problem with Relu is that it becomes inactive due to zero gradient for $x \lt 0$. Leaky Relu allows a small, positive grdient when the unit is not active.

```python
def leakyRelu(x):
  return (x if x > 0 else 0.01*x)

xVal = np.linspace(-5, 3, 200)
yRelu = [relu(x) for x in xVal]
yLRelu = [leakyRelu(x) for x in xVal]

fig = plt.figure(figsize=(6,6))
plt.plot(xVal, yRelu, '-', linewidth=3, label='Relu')
plt.plot(xVal, yLRelu, '-', linewidth=3, label='Leaky Relu')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x) = Leaky ReLU(x)$', fontsize=18)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# $\color{green}{\text{Sigmoid Linear Unit (SiLU)}}$
This is anoter variant of the ReLU activation function. It is also called $\color{blue}{\text{swish function}}$.
If $x$ is the input to SiLU neuron, the output is given by \\
$$f(x) = x \cdot \text{sigmoid(x)}$$


```python
def sigmoid(x):
  sigma = 2.0
  return (1/(1+np.exp(-sigma*x)))

def Silu(x):
  return (x * sigmoid(x))

def relu(x):
  return np.maximum(0.0, x)

xVal = np.linspace(-3, 3, 200)
yRelu = [relu(x) for x in xVal]
ySilu = [Silu(x) for x in xVal]

fig = plt.figure(figsize=(6,6))
plt.plot(xVal, yRelu, '-', linewidth=3, label='Relu')
plt.plot(xVal, ySilu, '-', linewidth=3, label='Silu')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# $\color{green}{\text{Gussian Error Linear Unit (GELU)}}$
This is smooth variant of the ReLU activation function. It has a non-monotonic "bump" for negative values of x.
If $x$ is the input to GELU neuron, the output is given by
$$f(x) = x \cdot \Phi(x)$$
where $\Phi(x)$ is the cumulative distribution function of the standard normal (Gaussian distribution) and is given by:
$$\Phi(x) = \frac{1}{\sqrt{2\pi}}\int_{-\infty}^{x}e^{-z^2/2}dz $$
To determine $\Phi(x)$, we can use Python's Math library function error function Erf$(x)$ given by
$$ \text{Erf}(x) = \frac{2}{\sqrt{2\pi}} \int_{0}^{x\sqrt{2}} e^{-z^2/2}dz$$
Therefore,
$$\Phi(x) = \frac{1+\text{Erf}(x/\sqrt{2})}{2}$$

```python
from math import erf, sqrt
def Phi(x):
  return (1.0 + erf(x/sqrt(2.0)))/2.0

def Gelu(x):
  return (x * Phi(x))

def relu(x):
  return np.maximum(0.0, x)

xVal = np.linspace(-3, 3, 200)
yRelu = [relu(x) for x in xVal]
yGelu = [Gelu(x) for x in xVal]

fig = plt.figure(figsize=(6,6))
plt.plot(xVal, yRelu, '-', linewidth=3, label='Relu')
plt.plot(xVal, yGelu, '-', linewidth=3, label='Gelu')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x) = GELU(x)$', fontsize=18)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# $\color{green}{\text{Softplus or SmoothReLU}}$
This is another smooth variant of the ReLU activation function.
If $x$ is the input to softplus neuron, the output is given by
$$f(x) = ln(1+e^x)$$
For large negative $x$, sofplus is slightly above $0$ and for large positive $x$, it is just above $x$. A more general form of this activation function is the one with a sharpness parameter $k$:
$$ f(x) = \frac{ln(1+e^{kx})}{k}$$
It should be noted that the derivative of the softplus activation function is the logistic function (which in turn is the smooth approximation of the binary sigmoid (Heaviside) function, that is the derivative of ReLU):
$$ f^{\prime}(x) = \frac{e^{kx}}{1+e^{kx}}=\frac{1}{1+e^{-kx}}$$

```python
from math import exp, log

def softplus(x):
  k = 5.0
  return log(1+exp(k*x))/k

def relu(x):
  return np.maximum(0.0, x)

xVal = np.linspace(-3, 3, 200)
yRelu = [relu(x) for x in xVal]
ySoftplus = [softplus(x) for x in xVal]

fig = plt.figure(figsize=(6,6))
plt.plot(xVal, yRelu, '-', linewidth=3, label='Relu')
plt.plot(xVal, ySoftplus, '-', linewidth=3, label='SoftPlus')
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$f(x)$', fontsize=18)
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
